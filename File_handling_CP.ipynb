```python
import numpy as np

# Open the CSV file and read its contents
file_path = '/content/Loan_prediction_dataset.csv'
with open("/content/Loan_prediction_dataset.csv", 'r') as file:
    data_type = [('Load_ID', 'U8'), ('Gender', 'U8'), ('Married', 'U8'), ('Dependents', np.int16), ('Education', 'U8'), ('Self_Employed', 'U8'), ('ApplicantIncome', np.int16), ('CoapplicantIncome', np.int16), ('LoanAmount', np.int16), ('Loan_Amount_Term', np.int16), ('Credit_History', np.int16), ('Property_Area', 'U8'), ('Loan_Status', 'U8')]
cp = np.genfromtxt('/content/Loan_prediction_dataset.csv', delimiter=',', dtype=data_type)



# Calculate mean, median, and standard deviation of loan amounts
mean_loan_amount = np.mean(cp['LoanAmount'])
print(f'mean_loan_amount: {mean_loan_amount}')

median_loan_amount = np.median(cp['LoanAmount'])
print(f'median_loan_amount: {median_loan_amount}')

std_loan_amount = np.std(cp['LoanAmount'])
print(f'std_loan_amount: {std_loan_amount}')
```
